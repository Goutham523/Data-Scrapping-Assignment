## Importing required libraries

```python
# imported all the required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# create a dataframe with given excel file
df1=pd.read_excel(r"C:\Users\gouth\Downloads\Web Scraping Assignment.xlsx")
df1.head()
```

## Dropping unnessary columns

```python
# can be drop the adress column
df1.drop("ADRESS", axis=1, inplace=True)
```

```python
df1 = df1.iloc[ :,1: ]
```

```python
df1.head()
```

```python
# filling null values with empty space

df1.fillna(" ", inplace = True)
```

```python
# we can extract the data from a single site given

page= requests.get("http://www.useams.org/")
soup = BeautifulSoup(page.content, "html.parser")
```

```python
# address of that college is

rep = soup.find('div', id = "biola-address")
print(rep.get_text())
```

```python
# we can copy the url links to another dataframe

URLS = df1.iloc[:,-1]
```

```python
URLS.head()
```

## Extracting the address of colleges from given links

```python
# firstly,create a list called 'addresses'
addresses = []

for url in URLS:
    try:
        # get the HTML content of the page
        if not url.startswith('http'):
            url = 'http://' + url
            
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        
            
        # find the address element on the page
        address_element = soup.find('address')
        
        if address_element:
            address = address_element.text.strip()
        else:
            address = ''
            
        addresses.append(address)
        print(f"{url}: Address - {address}")
        
    except requests.exceptions.ConnectionError:
        # handle the ConnectionError exception and skip the URL
        print(f"Unable to access {url}")
        addresses.append('')
```

```python
# add the list 'addresses' into df1 as column name 'Address'

df1['Address'] = addresses
```
